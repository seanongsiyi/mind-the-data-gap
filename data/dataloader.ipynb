```python
import pandas as pd
import duckdb
import os
```

```python
# Insert your local path to the data folder containing all datasets
DATA_DIR = r"C:\Users\shaog\OneDrive\Documents\DSE3101\data"
```

```python
# Creates a DuckDB database file
con = duckdb.connect(os.path.join(DATA_DIR, "..", "mind-the-data-gap", "data.duckdb"))
```

```python
# Create the database with 6 tables
con.execute(f"""
CREATE OR REPLACE TABLE abt_pt_jrny AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_abt_pt_jrny (full).csv')
""")

con.execute(f"""
CREATE OR REPLACE TABLE abt_pt_ride AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_abt_pt_ride (full).csv')
""")

con.execute(f"""
CREATE OR REPLACE TABLE pt_jrny AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_pt_jrny_all (full).csv')
""")

con.execute(f"""
CREATE OR REPLACE TABLE pt_ride AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_pt_ride_all (full).csv')
""")

con.execute(f"""
CREATE OR REPLACE TABLE bus_stop AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_q_bus_stop (full).csv')
""")

con.execute(f"""
CREATE OR REPLACE TABLE vls_marker AS
SELECT * FROM read_csv_auto('{DATA_DIR}/A0008D - v_q_vls_marker (full).csv')
""")
```

```python
con.close()
```
